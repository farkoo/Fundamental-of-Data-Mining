# <div dir="rtl">قسمت d:</div>

<div dir="rtl">

در خط اول یک mask ساختم که به ازای سطرهایی که شرایط خواسته شده را دارند مقدار True دارد و به ازای سطرهایی که این شرایط را ندارند مقدار False میگیرد.

سپس با استفاده از این ماسک داده ها را از دیتافریم اصلی استخراج میکنم و داخل دیتافریم df1 میریزم.

سپس ستونهایی که خواسته نشده را با استفاده از تابع drop حذف میکنیم.

سپس این دیتافریم جدید را با استفده از دستور to_csv بصورت فایل csv ذخیره میکنم.

نهایتا ده سطر آخر دیتافریم نهايی را با استفاده از تابع tail نمایش می‌دهیم.

<div>

```python
mask = (machine_data['MMIN'] > 1000) & (machine_data['MMIN'] <= 20000)
df1 = machine_data.loc[mask]
df1 = df1.drop(["MCYT", "CACH", "CHMIN", "CHMAX"], axis=1)
df1.to_csv("df1.csv")
```

```python
df1.tail(10)
```
