# <div dir="rtl">قسمت d:</div>

<div dir="rtl">

در این قسمت برای هر نمودار لازم است ویژگی‌ها را گروه بندی کنیم و تعداد هر گروه را برای هر ویژگی بدست آوریم و براساس این اطلاعات نمودار pie آن را رسم کنیم.

در اینجا با استفاده از تابع groupby هر ستون را گروه بندی کرده ام و با استفادع از تابع size تعداد اعضای هر گروه را بدست آوردم.

سپس با استفاده از تابع plot و با مشخص کردن نوع نموداری که میخواهم برای هر ستون این نمودار را رسم کرده ام.

از تابعی به نام label_function استفاده کردم تا در هر تکه از نمودار pie مشخص کنم چند درصد از داده ها به این قسمت اختصاص دارد.

<div>

```python
def label_function(val):
    return f'{val / 100 * len(student_data):.0f}\n{val:.0f}%'

fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(ncols=6, figsize=(60, 20))

student_data.groupby('sex').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 20},
                                  colors=['tomato', 'yellow'], ax=ax1)
student_data.groupby('internet').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 20},
                                  colors=['gold', 'skyblue'], ax=ax2)
student_data.groupby('address').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 20},
                                  colors=['violet', 'lime'], ax=ax3)
student_data.groupby('Pstatus').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 20},
                                  colors=['red', 'skyblue'], ax=ax4)
student_data.groupby('reason').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 20},
                                  colors=['violet', 'gold', 'skyblue'], ax=ax5)
student_data.groupby('schoolsup').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 20},
                                  colors=['green', 'tomato'], ax=ax6)

ax1.set_ylabel('Gender', size=40)
ax2.set_ylabel('Internet', size=40)
ax3.set_ylabel('Address', size=40)
ax4.set_ylabel('Pstatus', size=40)
ax5.set_ylabel('Reason', size=40)
ax6.set_ylabel('Schoolsup', size=40)

plt.show()
```

# <div dir="rtl">قسمت g:</div>

<div dir="rtl">

در این سوال خواسته شده نمرات را که داده ای عددی است به صورت categorical در بیاوریم و به 4 دسته تقسیم کنیم.

من این کار را بوسیله ی تابع cut انجام داده ام. حدود نمرات هر دسته را به این تابع داده ام و نام گروه ها را هم مشخص کردم و این داده های جدید را به دیتافریم اضافه کرده ام.

سپس در سلول بعدی به روشی که پیشتر نیز گفته شده داده های این ستون را دسته بندی کردم و بعد از استخراج تعداد هر گروه نمودار pie آن را رسم کردم.

گروه های مختلف با رنگبندی متفاوت نشان داده شده اند همچنین نام هر گروه و درصد فراوانی آن دسته در نمودار مشخص شده است.

<div>

```python
category = pd.cut(student_data.Grade,bins=[0,5,10,15,20],labels=['D','C','B','A'])
student_data.insert(student_data.shape[1],'Grade group',category)
```

```python
fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 5))
student_data.groupby('Grade group').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 8},
                                  colors=['red', 'yellow', 'skyblue', 'lime'], ax=ax1)
ax1.set_ylabel('Grade group', size=15)

plt.show()
```

# <div dir="rtl">قسمت h:</div>

<div dir="rtl">

در این قسمت خواسته شده که همبستگی داده ها را با استفاده از نمودار heatmap نمایش دهیم.

با استفاده از تابع corr از کتابخانه pandas ابتدا همبستگی ستونهای مختلف را بدست می‌آوریم. سپس با استفاده از مششخص کردن یک سطح آستانه و به کمک تابع heatmap از کتابخانه seaborn نمودار heatmap ویژگیها را نصب میکنیم.

بیشترین همبستگی بین ستونهای G2 و Grade است که با یکدیگر همبستگی 0.97 دارند.

<div>

```python
corr_matrix = student_data.corr().abs()

threshold = 0.01

filtered_corr_df = corr_matrix[(corr_matrix >= threshold) & (corr_matrix != 1.000)] 

plt.figure(figsize=(30,10))
sns.heatmap(filtered_corr_df, annot=True, cmap="Reds")
plt.show()
```
