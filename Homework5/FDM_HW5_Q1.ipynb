```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

# c.

```python
df['STORE_BILL_ID'] = df['STORECODE'] + '_' + df['BILL_ID']
```

```python
df['STORE_BILL_ID']
```

# d.

```python
sum_df = df.groupby(['STORE_BILL_ID','GRP'])['QTY'].agg({'sum'})
```

```python
sum_df
```

# e.

```python
df.QTY[df['QTY']>=1] = 1
df.QTY[df['QTY']<1] = 0
```

# f.

```python
df.groupby(["STORE_BILL_ID"])["GRP"].apply(list)
```

```python
te = TransactionEncoder()
te_ary = te.fit_transform(df.groupby(["STORE_BILL_ID"])["GRP"].apply(list))
```

```python
new_df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
new_df
```

```python
frequent_itemsets2 = apriori(new_df, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets2
```

# g.

```python
rules = association_rules(frequent_itemsets2, metric="lift")
```

```python
rules
```

<div style="direction:rtl">

conviction بالا به این معنی است که نتیجه به شدت به antecedent بستگی دارد. به عنوان مثال، در مورد یک نمره اطمینان کامل، مخرج 0 می شود (به دلیل 1 - 1) که برای آن نمره conviction به عنوان 'inf' تعریف می شود. مشابه lift، اگر موارد independent باشند، conviction 1 است.
    
</div>

<div style="direction:rtl">

Leverage تفاوت بین فرکانس مشاهده‌شده A و C که با هم ظاهر می‌شوند و فرکانسی که اگر A و C independent بودند را محاسبه می‌کند. مقدار اهرم 0 نشان دهنده استقلال است.
    
</div>

<div style="direction:rtl">

مثلا قاعده آخر مشخص میکند که پس از آمدن washing powders/liquids کالای tooth pastes خریداری میشود.
    
ساپورت قاعده ی پیشین برابر با 0.07 و ساپورت قاعده پسین برابر با 0.04 است
    
ساپورت و کانفیدنس قاعده برابر با 0.01 و 0.14 است.
    
مقدار lift برابر با 3.14 مقدار liverage برابر با 0.007 و مقدار conviction برابر با 1.11 میباشد.
    
</div>
